# LINE GRAPH(MATPLOTLIB)


```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

```python
x=[1,2,3,4]
```

```python
y=[5,6,7,8]
```

```python
plt.plot(x,y,'b',marker='o',linewidth="2",label="First")
a=[6,7,8,9]
b=[10,20,30,40]
plt.plot(a,b,'r',linestyle='--',marker='.',linewidth="3",label="Second")
plt.title("SAMPLE")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.legend()
plt.grid(True)
plt.show()
```

# Data Analysis

```python
r=[1,2,3,4]
```

```python
label=["one","two","three","four"]
```

```python
colors=['red','green','purple','blue']
```

```python
explode=[0.4,0,0,0]
```

```python
plt.pie(r,labels=label,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%',startangle=85,wedgeprops={'edgecolor':'black'})
```
